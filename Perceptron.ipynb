```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import kagglehub
import os
```

```python
# Download latest version
path = kagglehub.dataset_download("sameerprogrammer/college-placement")

print("Path to dataset files:", path)
```

```python
dataset_path = '/root/.cache/kagglehub/datasets/sameerprogrammer/college-placement/versions/1'
print(os.listdir(dataset_path))
```

```python
df = pd.read_csv(dataset_path + '/placement-dataset.csv')
```

```python
df.head()
```

```python
sns.scatterplot( x='cgpa', y='iq', hue = 'placement', data = df)
plt.title('Placement vs CGPA vs IQ')
plt.show()
```

```python
x = df.iloc[:, 2:4]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p  = Perceptron()
```

```python
p.fit(x,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values, y.values, clf=p, legend=2)
```
